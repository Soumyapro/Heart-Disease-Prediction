```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("framingham.csv")
df.head(10)
```

```python
df.describe()
```

```python
#Drop education column because there is no use of it in prediction
df = df.drop(columns=['education'])
```

```python
#now we will check for missing values
df.isna().sum()
```

```python
df.shape
```

```python
#value distribution in label.
df['TenYearCHD'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

columns_missing_values = ["cigsPerDay","BPMeds","totChol","BMI","heartRate","glucose"]
for col in columns_missing_values:
    plt.figure(figsize=(20,8))
    plt.subplot(1,2,1)
    sns.histplot(df[col],kde=True)
    plt.title(f"{col} distribution")

    plt.subplot(1,2,2)
    sns.boxplot(x=df[col])
    plt.title(f"{col} boxplot")

    plt.show()
    
```

```python
skewness = df[columns_missing_values].skew()
print(skewness)
```

```python
#Handling missing values
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")
for col in columns_missing_values:
    df[col] = imputer.fit_transform(df[[col]])
```

```python
#Now There is no missing values.
df.isna().sum()
```

```python
correlation_matrix = df.corr()
#visualize.
plt.figure(figsize=(20,8))
sns.heatmap(correlation_matrix,annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title("Correlation")
plt.show()
```

```python
X = np.asarray(df[["age","totChol","glucose","sysBP","prevalentHyp"]])
y = np.asarray(df["TenYearCHD"])
```

```python
#Feature Scaling and Feature splitting
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
X = preprocessing.StandardScaler().fit(X).transform(X)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
#Shapes
print("X_train :",X_train.shape)
print("y_train :",y_train.shape)
print("X_test :",X_test.shape)
print("y_test :",y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(X_train,y_train)
y_pred = logistic.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred)
print("Accuracy score is : ",score)
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix,classification_report
cm = confusion_matrix(y_pred,y_test)
plt.figure(figsize=(10,4))
sns.heatmap(cm,annot=True,fmt="d",cmap='Blues')
plt.show()
```

```python
#Details
print("The details of the report are: ")
print(classification_report(y_pred,y_test))
```
